### Шаг 2. Предобработка данных

Лишних данных нет, но есть пропуски. Заполним их значениями False, т.к. апартаментов в принципе меньше и это значение, наиболее вероятно, просто не было указано для обычных квартир:

Также заменим пропуски в данных о количестве балконов нулями, т.к. вероятно, при отсутствии балконов пользователи просто не указывали количество и изменим тип данных на целочисленный:

```python
estate_data['balcony'] = estate_data['balcony'].fillna(0)
estate_data['balcony'] = estate_data['balcony'].astype('int')
estate_data['balcony'].value_counts()
```

Проверим наименования населённых пунктов:

```python
estate_data['locality_name'].unique()
```

В данных есть различие в написании слова "посёлок" (поселок). Приведём это к единому написанию:

```python
def processing_village (locality):
    words = []
    for word in locality.split():
        if word == 'посёлок':
            words.append('поселок')
        else:
            words.append(word)
    return " ".join(words)
estate_data['locality_name'] = estate_data['locality_name'].astype('str')
estate_data['locality_name'] = estate_data['locality_name'].apply(processing_village)
estate_data['locality_name'].unique()
```

Учитывая, что застройка обычно типовая, заполним пропуски в остальных параметров, но сначала получим новые признаки, возможно они будут коррелировать с высотой потолков

- заполнили пропуски в признаках апартаментов на False (is_apartment), как на более вероятный
- проверили данные о количестве балконов и заполнили пропуски нулями и привели к целочисленному типу (balcony)

### Шаг 3. Посчитаем и добавим в таблицу новые данные

Посчитаем и добавим в таблицу цену квадратного метра (price_per_seq_m). Рассчитаем её поделив last_price (цена на момент снятия с публикации) на total_area (площадь квартиры в квадратных метрах (м²))

```python
estate_data['price_per_seq_m'] = estate_data['last_price'] / estate_data['total_area']
```

Теперь добавим день недели (weekday), месяц (month) и год (year) публикации объявления.

```python
estate_data['weekday'] = estate_data['first_day_exposition'].dt.weekday
estate_data['month'] = estate_data['first_day_exposition'].dt.month
estate_data['year'] = estate_data['first_day_exposition'].dt.year
estate_data.head()
```

Добавим в таблицу этаж (floor_variant). 0 - первый, 1 - последний, 2 - другой. Пропущенные значения заменим на значение "другой", как на более вероятное.

```python
#сделаем функцию обработки строки
def floor_var(row):
    if row['floor'] == 1:
        return 0
    elif row['floor'] == row['floors_total']:
        return 1
    else:
        return 2

estate_data['floor_variant'] = estate_data.apply(floor_var, axis=1)
estate_data.head(10)
```

Теперь заполним пропуски в данных о высоте потолков. Посмотрим с чем коррелирует высота потолков:

```python
plt.figure(figsize=(25, 15));
sns.heatmap(estate_data.corr(), annot=True)
```

Связанные признаки: апартаменты, общая площадь, удалённость от центра, цена за квадратный метр

Добавим в таблицу временные столбцы с категориями общей площади, удаленностью от центра и ценой за квадратный метр, где 0 будет для первого квантиля, 1 - для второго, 2 - для третьего, 3 - для четвертого

```python
quantiles_area = estate_data['total_area'].quantile(q=[0.25, 0.50, 0.75])
quantiles_area = quantiles_area.to_dict()

quantiles_cityCenters_nearest = estate_data['cityCenters_nearest'].quantile(q=[0.25, 0.50, 0.75])
quantiles_cityCenters_nearest = quantiles_cityCenters_nearest.to_dict()

quantiles_price_per_seq_m = estate_data['price_per_seq_m'].quantile(q=[0.25, 0.50, 0.75])
quantiles_price_per_seq_m = quantiles_price_per_seq_m.to_dict()
```

```python
def grouping(val, quantiles):
    if val <= quantiles[0.25]:
        return 0
    elif val <= quantiles[0.5]:
        return 1
    elif val <= quantiles[0.75]:
        return 2
    else:
        return 3
estate_data['group_total_area'] = estate_data['total_area'].apply(grouping, args=[quantiles_area])
estate_data['group_cityCenters_nearest'] = estate_data['cityCenters_nearest'].apply(grouping, args=[quantiles_cityCenters_nearest])
estate_data['group_price_per_seq_m'] = estate_data['price_per_seq_m'].apply(grouping, args=[quantiles_price_per_seq_m])
#estate_data['group_total_area'] = estate_data['total_area'].apply(lambda x: grouping(x, quantiles_area))
estate_data.head(10)
```

Посчитаем среднюю высоту потолка для каждого вида квартир:

```python
mean_ceiling_height = estate_data.pivot_table(index=['is_apartment', 'group_total_area', 'group_cityCenters_nearest', 'group_price_per_seq_m'], values='ceiling_height', aggfunc='mean')
mean_ceiling_height.rename(columns=lambda x: x.replace('ceiling_height', 'group_mean_ceiling_height'), inplace=True)
mean_ceiling_height.head(20)
```

А теперь заполним пропуски:

Добавим столбец со средним для группы:

```python
estate_data = pd.merge(estate_data, mean_ceiling_height, how='left', on=['is_apartment', 'group_total_area', 'group_cityCenters_nearest', 'group_price_per_seq_m'])
estate_data.head(10)
```

По первым 10 строкам видно, что данный подход имел смысл. Теперь заменим пропуски на значения характерные для группы:

```python
def replace_na(df):
    out = []
    df = df.fillna(0)
    for index, row in df.iterrows():
        if row['ceiling_height'] != 0:
            out.append(row['ceiling_height'])
        else:
            out.append(row['group_mean_ceiling_height'].round(2))
    return out
estate_data['ceiling_height'] = replace_na(estate_data[['ceiling_height', 'group_mean_ceiling_height']])
estate_data = estate_data.drop(['group_total_area', 'group_cityCenters_nearest', 'group_price_per_seq_m', 'group_mean_ceiling_height'], axis=1)
estate_data.head(10)
```

Доавим соотношение жилой площали к общей (ratio_live_total) и площади кухни к общей (ratio_kitchen_total). <br>

Т.к. в датафрейме имеются имеются пропущенные значения в столбцах жилой площади и площади кухни, то для анализа данных признаков создадим отдельный датафрейм, удалив пропуски. Заполнение пропусков не добавит данным информативности (но внесёт искажения), а удаление из датафрейма строк с пропущенными значениями в данных столбцах уменьшит полезный объем данных для анализа по другим признакам.

```python
area_appartments = pd.DataFrame()
area_appartments['last_price'] = estate_data['last_price']
area_appartments['living_area'] = estate_data['living_area']
area_appartments['kitchen_area'] = estate_data['kitchen_area']
area_appartments['total_area'] = estate_data['total_area']
area_appartments['ratio_live_total'] = area_appartments['living_area'] / area_appartments['total_area']
area_appartments['ratio_kitchen_total'] = area_appartments['kitchen_area'] / area_appartments['total_area']
area_appartments = area_appartments.dropna(subset= ['ratio_live_total',  'ratio_kitchen_total'])
area_appartments.head()
```
